**파일이 잘 받아졌는지 확인**

```
import os

dev_path = "/content/drive/My Drive/Voxceleb/dev/wav"
test_path = "/content/drive/My Drive/Voxceleb/test/wav"
```

```
# print # of labels, # of utterances
num_dev_labels = os.listdir(dev_path)
num_dev_utts = 0
for dir_path, dir_names, filenames in os.walk(dev_path):
    num_dev_utts += len(filenames)

num_test_labels = os.listdir(test_path)
num_test_utts = 0
for dir_path, dir_names, filenames in os.walk(test_path):
    num_test_utts += len(filenames)

print("[ dev] {:>5} speakers, {:>7} utterances".format(len(num_dev_labels), num_dev_utts))
print("[test] {:>5} speakers, {:>7} utterances".format(len(num_test_labels), num_test_utts))
```
