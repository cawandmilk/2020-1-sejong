# **주요 내용**

*   Xception, Inception V3 구조 확인

```python
from tensorflow.keras.applications.xception import Xception, decode_predictions

xcep = Xception(input_shape=(299,299,3))
xcep.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(xcep, show_shapes = True, to_file = 'model.png')
```

```python
from tensorflow.keras.applications import InceptionV3

icep = InceptionV3(input_shape=(299,299,3))
icep.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(icep, show_shapes = True, to_file = 'model.png')
```
